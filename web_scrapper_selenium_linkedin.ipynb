```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from IPython.display import display
import pandas as pd

import time
import datetime

result = []
"""
data
(
position_title: String, -> h2.top-card-layout__title
company: String, -> a.topcard__org-name-link
location: String, -> span.topcard__flavor.topcard__flavor--bullet
posted_time_ago: String, -> span.posted-time-ago__text
seniority_level: String, -> h3.description__job-criteria-subheader, span.description__job-criteria-text
employ_type: String,
job_function: String,
Industries: String
)

"""


options = Options()
options.add_argument("-headless")
driver = webdriver.Firefox(options=options)
start_url = "https://www.linkedin.com/jobs/search?trk=guest_homepage-basic_guest_nav_menu_jobs&position=1&pageNum=0"
driver.get(start_url)
location_search = driver.find_element(by=By.ID, value='job-search-bar-location')
print(location_search)
location_search.clear()
location_search.send_keys("Banjarmasin")
print(location_search.get_attribute("value"))
location_search.send_keys(Keys.ENTER)
time.sleep(10)
list_job = driver.find_elements(by=By.XPATH, value='//ul[@class="jobs-search__results-list"]/li')
if(len(list_job)>0):
  try:
    for idx, el in enumerate(list_job):
      if(idx>5):
        break
      driver.execute_script("arguments[0].scrollIntoView(true);", el)
      el.click()
      time.sleep(5)
      position = driver.find_elements(By.CSS_SELECTOR, 'h2.top-card-layout__title')
      company = driver.find_elements(By.CSS_SELECTOR, 'a.topcard__org-name-link')
      location = driver.find_elements(By.CSS_SELECTOR, 'span.topcard__flavor.topcard__flavor--bullet')
      posted_time_ago = driver.find_elements(By.CSS_SELECTOR, 'span.posted-time-ago__text')
      type_job_header = driver.find_elements(By.CSS_SELECTOR, 'h3.description__job-criteria-subheader')
      type_job_desc = driver.find_elements(By.CSS_SELECTOR, 'span.description__job-criteria-text')
      temp = dict()
      temp["Datetime Scrapped"] = datetime.datetime.now()
      temp["Position Title"] = position[0].get_attribute("innerText")
      temp["Company"] = company[0].get_attribute("innerText")
      temp["Location"] = location[0].get_attribute("innerText")
      temp["Posted Time Ago"] = posted_time_ago[0].get_attribute("innerText")
      for idx, el in enumerate(type_job_header):
        temp[type_job_header[idx].get_attribute("innerText")] = type_job_desc[idx].get_attribute("innerText")
      result.append(temp)
  finally:
    print(result)
    df = pd.DataFrame(result)
    driver.quit()
    display(df)
```
